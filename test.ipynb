```python
import scipy.io
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
mat = scipy.io.loadmat('SPtest.mat')
```

mat.keys()

```python
fft_list=[]
pft_list=[]
fst_list=[]
pst_list=[]
ftt_list=[]
ptt_list=[]
ffot_list=[]
pfot_list=[]
for i in range(0,1000):
    fft_list.append(mat['DATsp'][0][0][i])
    pft_list.append(mat['DATsp'][0][1][i])
    fst_list.append(mat['DATsp'][1][0][i])
    pst_list.append(mat['DATsp'][1][1][i])
    ftt_list.append(mat['DATsp'][2][0][i])
    ptt_list.append(mat['DATsp'][2][1][i])
    ffot_list.append(mat['DATsp'][3][0][i])
    pfot_list.append(mat['DATsp'][3][1][i])
```

```python
fixation_first_trial=pd.DataFrame(data=fft_list)
fixation_first_trial['event']='fixation'
fixation_first_trial['trial']=1

pursuit_first_trial=pd.DataFrame(data=pft_list)
pursuit_first_trial['event']='pursuit'
pursuit_first_trial['trial']=1

fixation_second_trial=pd.DataFrame(data=fst_list)
fixation_second_trial['event']='fixation'
fixation_second_trial['trial']=2

pursuit_second_trial=pd.DataFrame(data=pst_list)
pursuit_second_trial['event']='pursuit'
pursuit_second_trial['trial']=2

fixation_third_trial=pd.DataFrame(data=ftt_list)
fixation_third_trial['event']='fixation'
fixation_third_trial['trial']=3

pursuit_third_trial=pd.DataFrame(data=ptt_list)
pursuit_third_trial['event']='pursuit'
pursuit_third_trial['trial']=3

fixation_fourth_trial=pd.DataFrame(data=ffot_list)
fixation_fourth_trial['event']='fixation'
fixation_fourth_trial['trial']=4

pursuit_fourth_trial=pd.DataFrame(data=pfot_list)
pursuit_fourth_trial['event']='pursuit'
pursuit_fourth_trial['trial']=4
```

```python
df=fixation_first_trial.append([pursuit_first_trial, fixation_second_trial, pursuit_second_trial, fixation_third_trial, pursuit_third_trial, fixation_fourth_trial, pursuit_fourth_trial], ignore_index=True)
```
